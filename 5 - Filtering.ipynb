```python
# Import the color module
from skimage import color

# Import the filters module and sobel function
from skimage.filters import sobel

#import data
from skimage import data
camera = data.camera()

# Make the image grayscale
camera_image_gray = color.rgb2gray(camera)

# Apply edge detection filter
edge_sobel = sobel(camera_image_gray)

# Show original and resulting image to compare
from common import plot_comparison 
plot_comparison(camera,edge_sobel,"Edges with Sobel")
```

```python
# Import Gaussian filter 
from skimage.filters import gaussian

# Apply filter
gaussian_image = gaussian(camera,multichannel=True)

#plot
plot_comparison(camera,gaussian_image,"Reduced sharpness Gaussian")
```
